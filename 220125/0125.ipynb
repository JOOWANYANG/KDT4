# merge/ concat
* dataframe 을 합치는것
* `merge`는 sql의 table을 합하는 `join`과 같다

```python
# merge의 종류
# 두 dataframe 에 공통 칼럼이 존재해야 하고, 공통 칼럼을 기준으로 병합.

# A_df와 B_df를 병합하는 경우
# 1. inner merge : A_df 와 B_df 모두에 존재하는 데이터(row)들로만 합하는 방식 (디폴트)
# 2. outer merge : A_df 와 B_df 어느 한 곳에 존재하는 데이터는 모두 포함해서 합하는 방식
# 3. left merge  : A_df 의 데이터를 모두 포함. B_df에서 A_df에 존재하는 데이터만 포함
# 4. right merge : B_df 의 데이터를 모두 포함. A_df에서 B_df에 존재하는 데이터만 포함
```

```python
raw_data = {
    'subject_id':['1','2','3','4','5','7','8','9','10','11'],
    'test_score':[51,15,15,61,16,14,15,15,16,61]
}

df_a = pd.DataFrame(raw_data)
df_a
```

```python
raw_data = {
    'subject_id':['4','5','6','7','8'],
    'first_name':['Billy','Brain','Bran','Bryce','Betty'],
    'last_name':['Bonder','Black','Balwner','Brice','Btisan']
}

df_b = pd.DataFrame(raw_data)
df_b
```

```python
df_a.head(2)
```

```python
df_b.head(2) # 꼭 이름의 같아야 하는게 아니라 실질적으로 명시할 이름이 같아야 함.
```

```python
df_a.subject_id.values
```

```python
df_b.subject_id.values
```

```python
merged_df = pd.merge(df_a,df_b, on='subject_id', how='inner')
merged_df
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer') # outmerge : nan값 생성 가능성이 크다.
```

```python
df_a.merge(df_b , on='subject_id' , how='right') # 이렇게 해도 괜찮음
```

```python
df_a.merge(df_b , on='subject_id' , how='left') 
```

```python
pd.merge(df_a,df_b ,on='subject_id',how='right')
```
